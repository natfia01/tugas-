```
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import bernoulli
x = bernoulli.rvs(size=2000,p=0.55)
ax = sb.distplot(x,kde=True,color='black',hist_kws={"linewidth":25,'alpha':1})
plt.axhline(0.55,2000)
plt.xlabel("jumlah lemparan")
plt.ylabel("peluang lemparan")
plt.title("KEBERHASILAN LEMPARAN")
plt.show()
```

```
from scipy.stats import poisson
import matplotlib.pyplot as plt
mu = 5000
x = np.arange(2400,2700,0.51)
plt.plot(x,poisson.pmf(x,mu))
```
